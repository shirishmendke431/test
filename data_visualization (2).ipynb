```python
import  pandas as  pd
import  matplotlib.pyplot as  plt
import  seaborn as  sns
```

```python
path  = '/content/laptopPrice.csv'
df = pd.read_csv(path)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# drop  duplicates
df.drop_duplicates(keep  = 'first' ,  ignore_index  =  True  , inplace  = True)
```

```python
df
```

```python
df.dtypes
```

```python
#  use  count  plot for categorical data
```

```python
cat =  df.columns[df.dtypes == 'object']
num =  df.columns[df.dtypes  != 'object']
```

```python
#  plot countplot
plt.figure(figsize  =(10  ,  6))
sns.countplot(data  =  df , x= 'brand'  , color = 'orange' , )
plt.title("Count  plot  of  Brand")
plt.grid()
```

```python
#  for  loop
for  i in  cat :
  plt.figure(figsize  =  (9 ,6))

  sns.countplot(data  =  df , x =  i)
  plt.xlabel(f'Countplot  of  {i}')
  plt.ylabel(f"count  of  {i}")
  plt.title(f"Count  plot {i}")
  plt.grid()
  plt.show()

```

```python
# plot  hisplot  for  numerical data
for  i  in num:
  plt.figure(figsize  =  (10  , 6))
  sns.histplot(data   = df  ,  x  =  i)
  plt.title(f'{i}')
  plt.grid()
  plt.show()
```

```python
for  i  in num:

  sns.histplot(data   = df  ,  x  =  i,  kde  =  True)  #  kde  =  kernel  density estimator
  plt.title(f'{i}')
  plt.grid()
  plt.show()
```

```python
# boxplot -  used to detect  outliers
# outliers  -  observations  which are  far away  from  remaining  obsns
sns.boxplot(data  = df  ,  y =  'Price')
```

```python
# # lower  limit  =  q1 - 1.5*(q3-q1)
# #  upper  limit  =  q3 + 1.5(q3-q1)
# lower  limit  =  mu-3*sigma    # for normal   distribution  for  detection  of  outliers
# upper  limit  =  mu+3*sigma
# mu  = mean
# sigma  =  std  deviation
```
